```python
df['year'] = df['date_added'].dropna().apply(lambda x:x.split()[-1])
df['month'] = df['date_added'].dropna().apply(lambda x:x.split()[0])
month_order = ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', \
               'September', 'October', 'November', 'December']
```

```python
month_count = df.groupby('year')['month'].value_counts().unstack().fillna(0)[month_order].T
```

```python
fig = plt.figure(figsize=(12,8))
plt.title('Content Addition')
ax = sns.heatmap(month_count, cmap='gist_heat_r',edgecolors='white',linewidths=2)
```

```python
df2 = pd.DataFrame()
```

```python
df2['type'] = df['type']
df2['title'] = df['title']
df2['director'] = df['director']
df2['cast'] = df['cast']
df2['rating'] = df['rating']
```

```python
df2.head()
```

```python
df2 = df2.fillna(' ')
```

```python
for i in df2.columns:
    df2[i] = df2[i].apply(str.lower)
```

```python
df2.head()
```

```python
for i in range(len(df2)):
    df2.at[i,'bow'] = df2.loc[i,'title']+df2.loc[i,'director']+df2.loc[i,'cast']+df2.loc[i,'rating']
```

```python
count = CountVectorizer(stop_words='english')
count_matrix = count.fit_transform(df2['bow'])
```

```python
count_matrix
```

```python
cos_sim = cosine_similarity(count_matrix,count_matrix)
```

```python
df2 = df2.set_index('title',drop=True)
```

```python
indices = pd.Series(df2.index)
indices
```

```python
def recommend(Title,sim = cos_sim):
    recommended = []
    idx = indices[indices==Title].index[0]
    score_series = pd.Series(sim[idx]).sort_values(ascending=False)
    sim_index = list(score_series.iloc[1:11].index)
    for i in sim_index:
        recommended.append(df2.index[i])
    return recommended
```

```python
import pickle
```

```python
file = open('recommender_system.pkl','wb')
```

```python
pickle.dump(recommend,file)
```
